IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

READ CSV FILE

```python
iris=pd.read_csv('iris.csv')
iris.head()
```

HISTOGRAM

```python
iris.hist()
plt.show()
```

BOXPLOT

```python
iris.plot(kind="box")
plt.show()
```

SCATTERPLOT

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'])
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
plt.title('Scatter plot')
plt.show()
```

DISTRIBUTION CHART

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data=iris,x='sepal.length',kde=True,color='skyblue')
plt.xlabel("Sepal Length")
plt.show()
```

SCATTER MATRIX

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris,figsize=(10,10))
plt.show()
```

DENSITY CHART

```python
sns.kdeplot(data=iris,x='sepal.length',y='sepal.width',fill=True)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

ANDREWS CURVE

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10,10))
andrews_curves(iris,'variety',colormap='viridis')
plt.title('Andrews curve')
plt.show()
```

BUBBLE CHART

```python
plt.scatter(iris['sepal.length'],iris['sepal.width'],s=100,alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('sepal width')
plt.show()
```

PARALLEL CHART

```python
plt.figure(figsize=(10,6))
pd.plotting.parallel_coordinates(iris,'variety',colormap='viridis')
plt.show()
```

DEVIATION CHART

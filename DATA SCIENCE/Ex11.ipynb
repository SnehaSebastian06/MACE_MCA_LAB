```python
df1=pd.read_csv("/content/drive/MyDrive/Titanic_Train.csv")
df2=pd.read_csv("/content/drive/MyDrive/Titanic_Test.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df1['Sex'] = label_encoder.fit_transform(df1['Sex'])
df2['Sex'] = label_encoder.transform(df2['Sex'])

df1['Embarked'] = label_encoder.fit_transform(df1['Embarked'])
df2['Embarked'] = label_encoder.transform(df2['Embarked'])
```

```python
df1.fillna(0,inplace=True)
df2.fillna(0,inplace=True)
```

```python
df2.isnull().sum()
```

```python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Sex','Embarked']
X_train = df1[features]
y_train = df1['Survived']
X_test = df2[features]
```

```python
from sklearn.naive_bayes import GaussianNB
nb= GaussianNB()
nb.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_train_pred = nb.predict(X_train)
y_test_pred = nb.predict(X_test)
accuracy = accuracy_score(y_train, y_train_pred)
print(f"Accuracy on the training dataset: {accuracy}")
```
